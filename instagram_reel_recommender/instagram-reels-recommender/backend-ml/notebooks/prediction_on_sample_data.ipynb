```python
import pickle
import pandas as pd
import numpy as np
import sys
from sklearn.metrics import pairwise_distances
from sklearn.cluster import DBSCAN

sys.path.append("..")

preprocessing_pipeline = pickle.load(open("../full_pipeline.pkl", "rb"))
with open('cluster_centroids.pkl', 'rb') as f:
    cluster_centroids = pickle.load(f)
with open('cluster_labels.pkl', 'rb') as f:
    cluster_labels = pickle.load(f)
print("Preprocessor and cluster centroids loaded successfully.")
```

```python
old_df = pd.read_csv("../interaction_filtered.csv")
new_data = pd.DataFrame({
    "pid": [1],
    "category_level": [1],
    "author_fans_count": [46761],
    "watch_time": [136],
    "duration": [91.9],
    "cvm_like": [False],
    "comment": [False],
    "follow": [False],
    "collect": [False],
    "forward": [False],
    "hate": [False],
    "tag_name": ["正能量"],
    "gender": ["M"],
    "age": [42],
    "fre_city": ["邯郸"],
    "fre_community_type": ["unknown"],
    "fre_city_level": ["三线城市"]
})

transformed_data = preprocessing_pipeline.transform(new_data)
print("transformed_data after the PCA")
print(transformed_data)

cluster_centroids = pd.DataFrame(cluster_centroids)
```

```python
#finding the closest cluster.
distances = pairwise_distances(transformed_data , cluster_centroids.values)
closest_cluster_index = np.argmin(distances)
fallback_label = cluster_centroids.index[closest_cluster_index]
```
